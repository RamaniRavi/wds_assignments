## Arrays
**Given:** 1.5, 2.8, 0.2, 3.1

```python
import numpy as np

# Create array
x = np.array([1.5, 2.8, 0.2, 3.1])
print("Array:", x)

# Second element
print("Second element:", x[1])

# Elements 2 and 3
print("Elements 2 and 3:", x[1:3])

# Replace first element
x[0] = 9.9
print("After replacing first element:", x)

# Elements greater than 2
print("Elements > 2:", x[x > 2])

# Square values
x_squared = x**2
print("Squared values:", x_squared)

# Append 5.2
x = np.append(x, 5.2)
print("After appending 5.2:", x)

# Mean and length
print("Mean:", x.mean())
print("Length:", len(x))
```

## Lists

```python
# Create list
marks = [5, 7, 8, "Error", 5]
print("Marks list:", marks)

# Print Error
print("Error value:", marks[3])

# Show types
print("Types:", [type(i) for i in marks])

# Add marks
marks.extend([3, 5])
print("After adding marks:", marks)

# Replace Error with 0
marks[3] = 0
print("After replacing Error with 0:", marks)
```

## Dictionary

```python
# Initialize dictionary
medals = {}
print("Empty dictionary:", medals)

# Add values
medals = {"USA":112, "Germany":36, "China":100}
print("Initial medals:", medals)

# Add Italy
medals["Italy"] = 40
print("After adding Italy:", medals)

# Remove Germany
# del medals["Germany"]
medals.pop("Germany")
print("After removing Germany:", medals)

# Update USA
medals["USA"] = 113
print("After updating USA:", medals)

# Show keys and values
print("Keys:", medals.keys())
print("Values:", medals.values())
```

## Matrices

```python
# Save matrix
matrix = np.array([[1, 2],
                   [5, 4]])
print("Matrix:\n", matrix)

# Replace 5 with 0
matrix[matrix == 5] = 0
print("After replacing 5 with 0:\n", matrix)

# Sum first row
print("Sum of first row:", matrix[0].sum())

# Sum first column
print("Sum of first column:", matrix[:,0].sum())

# Row and column names
rows = ["Row1", "Row2"]
cols = ["Col1", "Col2"]
print("Row names:", rows)
print("Column names:", cols)
```
