## Arrays
**Given:** 1.5, 2.8, 0.2, 3.1

```python
import numpy as np

# Create array
x = np.array([1.5, 2.8, 0.2, 3.1])

# Second element
x[1]

# Elements 2 and 3
x[1:3]

# Replace first element
x[0] = 9.9

# Elements greater than 2
x[x > 2]

# Square values
x_squared = x**2

# Append 5.2
x = np.append(x, 5.2)

# Mean and length
x.mean(), len(x)
```

## Lists

```python
# Create list
marks = [5, 7, 8, "Error", 5]

# Print Error
marks[3]

# Show types
[type(i) for i in marks]

# Add marks
marks.extend([3, 5])

# Replace Error with 0
marks[3] = 0

marks
```

## Dictionary

```python
# Initialize dictionary
medals = {}

# Add values
medals = {"USA":112, "Germany":36, "China":100}

# Add Italy
medals["Italy"] = 40

# Remove Germany
medals.pop("Germany")

# Update USA
medals["USA"] = 113

# Show keys and values
medals.keys(), medals.values()
```
