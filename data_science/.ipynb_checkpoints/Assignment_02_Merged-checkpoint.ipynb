## Creating Data Frame

You are given a fragment of data about medals from the Olympics:

| Country | Gold | Silver | Bronze |
|---|---:|---:|---:|
| NOR | 16 | 8 | 13 |
| GER | 12 | 10 | 5 |
| CHN | 9 | 4 | 2 |

### Setup (imports)

```python
import pandas as pd
import numpy as np
```

### 1) Create a pandas DataFrame from this data

```python
df_medals = pd.DataFrame({
    "Country": ["NOR","GER","CHN"],
    "Gold": [16, 12, 9],
    "Silver": [8,10,4],
    "Bronze": [13,5,2]
})
df_medals
```

### 2) Select the first and third rows and the first and third columns

```python
df_medals.iloc[[0,2],[0,2]]
```

### (Optional) Select the 3rd row using integer position (your additional step)

```python
df_medals.iloc[2]
```

### 3) Return a data frame that contains just the CHN row

```python
df_medals[df_medals["Country"] == "CHN"]
```

### 4) Return a data frame that contains the medals of Norway and Germany

```python
df_medals[df_medals["Country"].isin(["NOR", "GER"])]
```

### 5) Select countries who won ten or more gold medals and return the Country and Gold columns

```python
df_medals.loc[df_medals["Gold"] >= 10, ["Country", "Gold"]]
```

### 6) Sort the medals by bronze (descending)

```python
df_medals.sort_values(by = "Bronze", ascending=False)
```
